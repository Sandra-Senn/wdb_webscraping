### Projektziel

- Automatisches Scrapen von Nachrichtenartikeln der Schweizer Radio und Fernsehen (SRF) Webseite.
- Extraktion strukturierter Metadaten (Titel, Datum, Autor, Kategorie, Unterkategorie) aus Artikeln.
- Speicherung der Daten in CSV-Dateien zur Weiterverarbeitung
- Explorative Datenanalyse (EDA) zur Erkennung von Mustern in den Kategorien je Suchbegriff


### Reale Anwendungen
- Medienbeobachtung: Identifikation der häufigsten Themen und Nennungen spezifischer Namen.
- Öffentliche Meinungsanalyse: Welche Kategorien werden mit bestimmten Namen oder Themen häufig assoziiert?
- Datenbeschaffung für NLP-Modelle: Bereitstellung gelabelter Textdaten für spätere Textklassifikation oder Named Entity Recognition.

### Datenstruktur & Speicherung
- Jeder Artikel wird als Dictionary mit folgenden Feldern gespeichert:
  - `Titel`: Titel des Artikels
  - `Datum`: Veröffentlichungsdatum
  - `Autor`: Verfasser, wenn verfügbar
  - `Kategorie und Unterkategorie`: Navigationsstruktur von SRF
  - `Suchbegriff`: Der im Scraping verwendete Begriff
- Alle Artikel werden in einem Pandas DataFrame gesammelt und als CSV-Datei unter ../output/df_scraping.csv gespeichert.


### Ablauf und Komponenten
1. init_driver(): Initialisiert den Selenium Webdriver mit Performance-Optimierungen (z. B. headless, keine Bilder, schnellerer Seitenaufbau).
2. load_frontpage(): Lädt die Suchergebnisseite und klickt wiederholt auf "Mehr anzeigen", bis genügend Artikel geladen sind.
3. process_article(): Öffnet jeden Artikel separat und extrahiert zusätzliche Metadaten.
4. get_article_info(): Führt den gesamten Extraktionsprozess pro Artikel aus.
5. scrape_srf_news(): Wrapper-Funktion, um Artikel für einen Begriff zu scrapen.
6. main(): Parallele Verarbeitung mehrerer Begriffe via ThreadPoolExecutor und CSV-Speicherung.


### Teststrategie
Die Tests im Skript bestehen aus folgenden Kontrollen und Sicherungen:
- Initialisierung des Drivers: Es wird getestet, ob der Driver richtig Initialisiert wird.
- Testen ohne Suchvariable: Der Scraper wird kurz ausgeführt ohne eine Suchvariable zu bekommen. Es wird geprüft, dass das Resultat eine leere Liste ist.
- Testen mit Suchvariable: Es wird einmal nach 2 Artikeln zu Thema Klima gescraped. Dabei wird überprüft, dass die Ausgabe überall Werte enthält.
Diese Test werden bei jedem Push auf Github automatisch durchgeführt, Fehler sind unter Actions aufzuginden.

Zudem haben wir den Code mit Loggings ausgestattet. Während des Scrapens werden die ausgeführten Aktionen in einer Logdatei gespeichert. Dies hilft um zu kontrollieren, ob der Code sauber läuft und kann helfen bei Problemen die Ursache zu finden.
